```python
tickersRowData = pandas.read_csv('NASDAQ-TickerList-20210922.csv')
tickers = tickersRowData['Symbol'].tolist()
```

```python
def check_stock_symbol():
    stock_symbol = input('choose stock symbol:')
    if stock_symbol in tickers:
        return stock_symbol
    else:
        return check_stock_symbol()
stock_symbol= check_stock_symbol()

set_cash = int(input('set cash:'))
set_strategy = input('slect strategy(BuyBuyBuy or Mystrategy ):')
from_Time = input('from_Time:')

if set_strategy=='Mystrategy':
  month_buy  = int(input('how long do you invest?(unit:one month):'))
  buy_size = int(input('how much can you invest at a time?:'))

nowTime = int(time.time()) # 取得現在時間
struct_time = time.localtime(nowTime) # 轉換成時間元組
nowTime  = time.strftime("%Y%m%d ", struct_time) # 將時間元組轉換成想要的字串

sy=int(from_Time[:4])
sm = int(from_Time[4:6])
sd = int(from_Time[6:8])
ey = int(nowTime[:4])
em = int(nowTime[4:6])
ed = int(nowTime[6:8]) #扣掉時差
```

```python
def download_csv():
  headers = {
      'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36',
  }
  s = requests.Session()

  start_timestamp = int(datetime.timestamp(datetime(sy,sm,sd)))
  end_timestamp = int(datetime.timestamp(datetime(ey,em,ed)))

  url = f'https://query1.finance.yahoo.com/v7/finance/download/{stock_symbol}?period1={start_timestamp}&period2={end_timestamp}&interval=1d&events=history&includeAdjustedClose=true'

  r = s.get(url, headers=headers)
  if r.status_code == requests.codes.ok:
      with open(f'/Users/xuziyuan/Desktop/finance/data/{stock_symbol}.csv', 'w') as trick_csv:
          trick_csv.write(r.text)

  if not os.path.isfile("/Users/xuziyuan/Desktop/finance/data/"+stock_symbol+".csv"):
      print('沒有此檔案')
  else:
      print(stock_symbol+'_'+'檔案成功下載')
```

```python
cerebro = bt.Cerebro()#大腦
```

```python
def read_data():
  data = bt.feeds.YahooFinanceCSVData(
      # 股票資料位置 './data/TSLA.csv'
      dataname="/Users/xuziyuan/Desktop/finance/data/"+stock_symbol+".csv",
      
      # 資料起始日
      fromdate=datetime(sy,sm,sd),
      
      # 資料結束日
      todate=datetime(ey,em,ed)
      
      )
  return data
```

```python
class Mystrategy(bt.Strategy):#多久買一股
     def __init__(self):
         self._next_buy_date = datetime(sy, sm, sd)
  
     def next(self):
         if self.data.datetime.date() >= self. _next_buy_date.date():
             self. _next_buy_date += relativedelta(months=month_buy)
             self.buy(size=buy_size)
class BuyBuyBuy(bt.Strategy):
  def log(self, txt, dt=None):
    ''' 買買買 策略 '''
    dt = dt or self.datas[0].datetime.date(0)
    print(f'{dt.isoformat()}, {txt}')

  def __init__(self):
    # 鎖定"收盤價"在 datas[0] 的收盤價
    self.dataclose = self.datas[0].close
    self.order = None #訂單的變數確保在next()每一次調用的時候，可以知道有沒有訂單的狀況
  #當訂單被放置時就會call notify_order()就會收到訂單資訊
  def notify_order(self, order):
    if order.status in [order.Submitted, order.Accepted]:
      return 
    if order.status in [order.Completed]:
      if order.isbuy():
        self.log(f'已購買於 {order.executed.price}')
      elif order.issell():
        self.log(f'已賣出於 {order.executed.price}')

      self.bar_executed = len(self)#總共有多少個K線

    self.order = None #只要訂單被放置就會把order清空

  def next(self):
    self.log(f'收盤價, {self.dataclose[0]}')#最新的收盤價是多少
    
    # 假設沒有倉位(股票資產)就會去買
    if not self.position:
      # 今日收盤價 < 昨日收盤價
      if self.dataclose[0] < self.dataclose[-1]:
        
        # 昨日收盤價 < 前日收盤價 兩個條件都符合才是連三天跌
        if self.dataclose[-1] < self.dataclose[-2]:
          self.log(f'購買信號, {self.dataclose[0]}')
          self.order = self.buy()
    # 否則就會去賣，self.bar_executed + 5在5個K線生成後就賣
    else:
      if len(self) >= (self.bar_executed + 3):
        self.log(f'賣出信號 {self.dataclose[0]}')
        self.order = self.sell()#把賣出的細節全部綁定到order裡面
```

```python
def choose_strategy():
        if set_strategy =='BuyBuyBuy':
            return cerebro.addstrategy(BuyBuyBuy)
        elif set_strategy =='Mystrategy':
            return cerebro.addstrategy(Mystrategy)
```

```python
def result_photo():
  cerebro.adddata(read_data()) 
  choose_strategy()
  # cerebro.addstrategy(bt.analyzers.DrawDown)
  cerebro.broker.set_cash(set_cash)
  cerebro.run()
  portvalue = cerebro.broker.getvalue()
  pnl = portvalue-set_cash
  #Print out the final result
  print('Final Portfolio Value: ${}'.format(portvalue))
  print('P/L: ${}'.format(pnl))
  
  # feedback = ((cerebro.broker.getvalue()/set_cash)-1)*100
  # print('{:.0f}'.format(feedback)+'%')
 
  cerebro.plot(style='candlestick') 
```
